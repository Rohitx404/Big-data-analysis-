# Superstore Profitability Analysis
This notebook performs:
- Data Import
- Data Cleaning
- Summary Statistics
- Visualization
- Profit Prediction using Linear Regression
- Business Interpretation


```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

## 1. Import Dataset

```python
df = pd.read_csv('dataset.csv')
df.head()
```

## 2. Explore Data

```python
df.info()
df.describe()
```

## 3. Data Cleaning

```python
# Remove missing values
df = df.dropna()

# Remove data entry errors
df = df[df['Sales'] > 0]
df = df[df['Quantity'] > 0]
```

## 4. Outlier Removal using IQR

```python
Q1 = df['Sales'].quantile(0.25)
Q3 = df['Sales'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df_clean = df[(df['Sales'] >= lower) & (df['Sales'] <= upper)]
df_clean.shape
```

## 5. Summary Statistics

```python
df_clean[['Sales','Profit','Discount','Quantity']].describe()
```

## 6. Visualization

```python
# Profit by Category (Bar Chart)
profit_category = df_clean.groupby('Category')['Profit'].sum()
profit_category.plot(kind='bar')
plt.xlabel('Category')
plt.ylabel('Total Profit')
plt.title('Profit by Product Category')
plt.show()
```

```python
# Scatter Plot Sales vs Profit
plt.scatter(df_clean['Sales'], df_clean['Profit'])
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.title('Sales vs Profit Relationship')
plt.show()
```

## 7. Machine Learning Model (Linear Regression)

```python
features = df_clean[['Sales','Discount','Quantity']]
target = df_clean['Profit']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    features, target, test_size=0.2, random_state=42
)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)

r2 = r2_score(y_test, predictions)
rmse = np.sqrt(mean_squared_error(y_test, predictions))

print('R2 Score:', r2)
print('RMSE:', rmse)
```

## 8. Feature Importance

```python
coefficients = pd.DataFrame({
    'Feature': features.columns,
    'Impact': model.coef_
})

coefficients
```

## Conclusion
- Sales strongly affects profit
- High discount can reduce profitability
- Companies should control discounts and focus on profitable categories
